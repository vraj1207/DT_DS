# DATA Visualization _ MATPLOTLIB

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a= np.array([1,2,3,4,5,6,7])
b= np.array([1,2,3,4,5,6,7])
```

```python
plt.title("Linear Line")
plt.xlabel("a")
plt.ylabel("b")
plt.plot(a,b)
plt.show()
```

```python
g= np.array([3,6,44,12,67,45,55])
h= np.array([22,4,5,6,78,22,12])
plt.xlabel('g')
plt.ylabel('h')
plt.plot(g,h)
plt.show()
```

```python
plt.plot(g,h,'--g')
plt.show()
```

```python
x=[5,8,10]
y=[12,16,6]

plt.bar(x,y)
plt.title("Bar Graph")
plt.ylabel('Y Axis')
plt.xlabel('X Axis')
plt.show()
```

```python
plt.bar(x,y,color="g",align="center")
plt.title("Bar Graph")
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
plt.show()
```

```python
#HISTOGRAM

a = np.array([22,3,44,55,66,77,88,99,23,32,43])

plt.hist(a,bins=[0,20,40,60,80,100])
plt.title("Histogram")
plt.show()
```
